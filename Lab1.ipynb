```python
with open("scores.csv", "w") as f:
    f.write(
        "Name,Score\n"
        "Andrew,88.3\n"
        "Ben,92.6\n"
        "Carol,89.7\n"
    )
```

```python
import pandas as pd
import math
```

```python
try:
    n = int(input("Enter n: "))

    with open("ages.txt", "r") as f:
        data = f.read().split(",")

    if len(data) < n:
        print("File contains fewer than n numbers.")

    ages = [math.floor(float(x)) for x in data[:n]]
    print("Ages list:", ages)

except Exception as e:
    print("Error reading ages:", e)
```

```python
try:
    df_scores = pd.read_csv("scores.csv")
    dic = df_scores.to_dict(orient="list")
    print(dic)
except Exception as e:
    print("Error reading scores:", e)
```

```python
try:
    dic["Age"] = ages
    df = pd.DataFrame(dic)
    print(df)
except Exception as e:
    print("Error creating DataFrame:", e)
```

```python
try:
    max_age = df["Age"].max()
    print("Maximum age:", max_age)
except Exception as e:
    print("Error finding maximum age:", e)
```

```python
try:
    df.loc[len(df)] = ["Daniel", 10, 78]
    print(df)
except Exception as e:
    print("Error appending row:", e)
```

```python
try:
    p = float(input("Enter deduction p: "))
    df.loc[1, "Score"] -= p
    print(df)
except Exception as e:
    print("Error applying deduction:", e)
```

```python
def round_half_up(x):
    return math.ceil(x * 2) / 2

try:
    df["Score"] = df["Score"].apply(round_half_up)
    print(df)
except Exception as e:
    print("Error rounding scores:", e)
```

```python
try:
    mean_score = df["Score"].mean()
    median_score = df["Score"].median()

    print("Updated DataFrame:")
    print(df)
    print("Mean Score:", mean_score)
    print("Median Score:", median_score)
except Exception as e:
    print("Error calculating statistics:", e)
```

```python
try:
    df_filtered = df[(df["Age"] >= 12) & (df["Age"] <= 15)]
    print("Filtered DataFrame:")
    print(df_filtered)
except Exception as e:
    print("Error filtering DataFrame:", e)
```

```python
try:
    mean_f = df_filtered["Score"].mean()
    median_f = df_filtered["Score"].median()

    print("Mean Score in Filtered DataFrame:", mean_f)
    print("Median Score:", median_f)
except Exception as e:
    print("Error calculating filtered statistics:", e)
```
